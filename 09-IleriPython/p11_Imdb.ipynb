```python
url = "https://www.imdb.com/chart/top/"

import requests
from bs4 import BeautifulSoup as bs

html = requests.get(url).content
#all html

soup = bs(html,"html.parser")
```

```python
list = soup.find("tbody",{"class":"lister-list"}).find_all("tr",limit=200)
count=0
for tr in list:
    title = tr.find("td",{"class":"titleColumn"}).find("a").text
    year = tr.find("td",{"class":"titleColumn"}).find("span").text.strip("()")
    raiting = tr.find("td",{"class":"ratingColumn imdbRating"}).find("strong").text
    count+=1
    print(f"{count}- Film: {title.ljust(50)} Yıl: {year} Değerlendirme: {raiting}")
```
