```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

x = np.linspace(-10,10,20)
y = x**3
z = x**2
figure = plt.figure()

axes = figure.add_axes([0.1,0.1,1,1])
axes.plot(x,y,"o-b") 

axes.set_xlabel("X axis")
axes.set_ylabel("Y axis")
axes.set_title("Cube")

axes1 = figure.add_axes([0.3,0.8,0.3,0.3])
axes1.plot(x,z,"o-r") 

axes1.set_xlabel("X axis")
axes1.set_ylabel("Z axis")
axes1.set_title("Square")


plt.show()
```

```python
figure = plt.figure()
axes1 = figure.add_axes([0,0,1,1])
axes1.plot(x,z,"o-r") 

axes1.set_xlabel("X axis")
axes1.set_ylabel("Z axis")
axes1.set_title("Square")

plt.show()
```

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(x,z,label="Square")
axes.plot(x,y,label="Cube")
axes.legend(loc=4)

plt.show()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,8))

axes[0].plot(x,y)
axes[0].set_title("Cube")


axes[1].plot(x,z)
axes[1].set_title("Square")
plt.tight_layout()

fig.savefig("figure1.png")
plt.show()
```
