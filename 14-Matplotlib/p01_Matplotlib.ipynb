```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x= np.arange(4)
y= [1,4,9,16]

plt.plot(x,y,"o--g")
plt.axis([0,5,0,15])

plt.title("Grafik")
plt.xlabel("X")
plt.ylabel("y")

plt.show()
```

```python
x = np.linspace(0,2,100)

plt.plot(x,x,label="linear",color="aqua")
plt.plot(x,x**2,label="quadratic",color="darkblue")
plt.plot(x,x**3,label="cubic",color="black")

plt.xlabel("x")
plt.ylabel("y")


plt.legend()
plt.show()

```

```python
fig,axs=plt.subplots(3)

axs[0].plot(x,x,color="aqua")
axs[0].set_title("Linear")

axs[1].plot(x,x**2,color="blue")
axs[1].set_title("Linear")

axs[2].plot(x,x**3,color="darkblue")
axs[2].set_title("Linear")

plt.tight_layout()
plt.show()
```

```python
fig,axs=plt.subplots(2,2)
fig.suptitle("Grafik bal")

axs[0,0].plot(x,x,color="aqua")
axs[0,0].set_title("Linear")

axs[0,1].plot(x,x**2,color="blue")
axs[0,1].set_title("quadratic")

axs[1,0].plot(x,x**3,color="darkblue")
axs[1,0].set_title("cubic")
4,
axs[1,1].plot(x,x**1000,color="black")
axs[1,1].set_title("n")

plt.tight_layout()
plt.show()
```
